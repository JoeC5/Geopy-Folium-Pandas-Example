Leveraging GeoPy, Folium, and Python I will be mapping firestations in my area.

Pull in geopy. And greate a dataset of firehouse locations.

```python
## import GeoPy
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="test_app")

## list of firestation locations - I pulled this from a Google search
addresses = ["3425 Issaquah Pine Lake Rd SE, Sammamish, WA",
             "1851 228th Ave NE, Sammamish, WA",
             "1280 NE Park Dr, Issaquah, WA",
             "1575 NW Maple St, Issaquah, WA",
             "16100 NE 8th St, Bellevue WA",
             "175 Newport Way NW, Issaquah, WA",
             "6502 185th Ave NE, Redmond, WA",
             "190 E Sunset Way, Issaquah, WA",
             "5021 264th Ave NE, Redmond, WA",
             "2802 148th Ave SE, Bellevue, WA",
             "8791 208th Ave NE, Redmond, WA",
             "22710 NE Alder Crest Dr, Redmond, WA",
             "4216 Factoria Blvd SE, Bellevue, WA",
             "5701 Lakemont Blvd SE, Bellevue, WA",
             "16917 NE 116th St, Redmond, WA",
             "12412 Newcastle Way, Bellevue, WA",
             "8450 161st Ave NE, Redmond, WA"
             ]
```

To leverage GeoPy we will need the logitude and latitude coordinates for each of the above locations. The following loop using geocode wil provide this information.

```python
##loop thru each address for latitude and longitude
##increased timeout to 10 secondsgiving the Nominatim server more time to resond
##not sure why the 8791 208th Ave NE, Redmond, WA location is not found. it is a valid location - I looked the location coordinates up.
##take these coordinates and copy/paste to an Excel spreasheet. The spreadsheet will be used to build the map
for address in addresses:
  location = geolocator.geocode(address, timeout=10)
  if location:
    print(f"{location.latitude}, {location.longitude}")
  else:
    print(f"{address} -> Location not found")
```

Build the map leveraging the data (from above) and Excel spreadsheet.

```python
## import pandas and create a dataframe
## I saved the excel spreadsheet into my Google Collab instance

import pandas as pd

df = pd.read_excel("/content/Fire_station_with_corrodinates.xlsx")
```

```python
## a quick information check of the data
df.info()
```

```python
import folium

# create a map centered around the first firestation in the dataframe
m = folium.Map(location=[df.iloc[0]['latitude'], df.iloc[0]['Longitude']], zoom_start=12)

#add markers for each firestation
for index, row in df.iterrows():
    folium.Marker([row['latitude'], row['Longitude']], popup=row['street address']).add_to(m)  ##allows the user to see actual address by clicking on marker
    folium.Circle(
        location=[row['latitude'], row['Longitude']],
        radius=2 * 1609.34, ##converting meters to miles - in this case 2 mile radius
        color='yellow',
        fill=True,
        fill_color='yellow'
    ).add_to(m)

#display the map
display(m)
```
